```python
import numpy as np
import numpy.linalg as npla
rng = np.random.default_rng(6174)
```

```python
for i in range(0, 10, 1):
 n = int(rng.integers(low=1, high=25, size=1))
 A = rng.random((8, n)) @ rng.normal(size=(n, 5))
 print("For matrix no. "+str(i)+":")
 print("Rank = {}".format(npla.matrix_rank(A)))
 print("AAt eigenvalues = {}".format(npla.eigvals(A @ np.matrix.transpose(A))))
 print("AtA eigenvalues = {}\n".format(npla.eigvals(np.matrix.transpose(A) @ A)))
```

Number of eigenvalues, i.e. rank of the matrix, is equal to size of the matrix in each case. When matrices are generated from random numbers, this is almost guaranteed to happen.

The numbers printed are not what are stored in the computer.

Some eigenvalues are complex. These are symmetric matrices and should not be having complex eigenvalues.
